```python
from sklearn import preprocessing
label_encoded = preprocessing.LabelEncoder()

def encode_column(column):
    label_to_number = column
    label_to_number = label_encoded.fit_transform(label_to_number)
    return label_to_number
```

```python
def add_column_to_df(columnName, df, valueColumn):
    df[columnName] = valueColumn
    return df1
```

```python
age_to_number = encode_column(df1.Grupo_edad)
age_to_number
```

```python
df1 = add_column_to_df('Age_number', df1, age_to_number)
df1
```

```python
income_to_number = encode_column(df1.Nivel_ingreso)
income_to_number
```

```python
df1 = add_column_to_df('Income_number', df1, income_to_number)
df1

```

```python
state_to_number = encode_column(df1.Entidad_Federativa)
state_to_number

df1 = add_column_to_df('State_number', df1, state_to_number)
df1
```

```python
sex_to_number = encode_column(df1.Sexo)
sex_to_number

df1 = add_column_to_df('Sex_number', df1, sex_to_number)
df1
```

```python
# Split independent vars(X) and dependent var(Y)
# here we have 2 variables for multiple regression. 
#If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.
#Alternatively, you may add additional variables within the brackets
X = df1[['Numero_personas','Age_number','State_number','Sex_number']] 
Y = df1['Income_number']
 
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)


# prediction with sklearn
Num_persons = 50000
Age_number = 4
State_number = 20
Sex_number = 0
print ('Predicted Income group : \n', regr.predict([[Num_persons, Age_number, State_number, Sex_number]]))
```

```python
# with statsmodels
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print(print_model)
```

```python
 tkinter GUI
root= tk.Tk() 
 
canvas1 = tk.Canvas(root, width = 1200, height = 450)
canvas1.pack()

# with sklearn
Intercept_result = ('Intercept: ', regr.intercept_)
label_Intercept = tk.Label(root, text=Intercept_result, justify = 'center')
canvas1.create_window(260, 220, window=label_Intercept)

# with sklearn
Coefficients_result  = ('Coefficients: ', regr.coef_)
label_Coefficients = tk.Label(root, text=Coefficients_result, justify = 'center')
canvas1.create_window(260, 240, window=label_Coefficients)

# with statsmodels
print_model = model.summary()
label_model = tk.Label(root, text=print_model, justify = 'center', relief = 'solid', bg='LightSkyBlue1')
canvas1.create_window(800, 220, window=label_model)


# New_Interest_Rate label and input box
label1 = tk.Label(root, text='Type Interest Rate: ')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)

# New_Unemployment_Rate label and input box
label2 = tk.Label(root, text=' Type Unemployment Rate: ')
canvas1.create_window(120, 120, window=label2)

entry2 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 120, window=entry2)


def values(): 
    global New_Interest_Rate #our 1st input variable
    New_Interest_Rate = float(entry1.get()) 
    
    global New_Unemployment_Rate #our 2nd input variable
    New_Unemployment_Rate = float(entry2.get()) 
    
    Prediction_result  = ('Predicted Stock Index Price: ', regr.predict([[New_Interest_Rate ,New_Unemployment_Rate]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 280, window=label_Prediction)
    
button1 = tk.Button (root, text='Predict Stock Index Price',command=values, bg='orange') # button to call the 'values' command above 
canvas1.create_window(270, 150, window=button1)
 

root.mainloop()
```
